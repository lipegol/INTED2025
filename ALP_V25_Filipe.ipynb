4 - Crie um algoritmo que receba uma idade do usuário, e depois verifique em qual faixa etária está a idade:

até 12 - Você é uma criança!

de 13 à 18 - Você é um adolescente!

de 19 à 60 - Você é um adulto!

acima de 60 - Você é um idoso!

```python
# Solicita a idade do usuário
idade = int(input("Digite sua idade: "))

# Verifica a faixa etária
if idade <= 12:
    print("Você é uma criança!")
elif 13 <= idade <= 18:
    print("Você é um adolescente!")
elif 19 <= idade <= 60:
    print("Você é um adulto!")
else:
    print("Você é um idoso!")
```

Peça um número de 1 a 7 e informe o dia da semana correspondente. Crie um algoritmo em Python que associe o número ao dia correto

```python
numero = int(input("Digite um número de 1 a 7: "))

dias = {
    1: "Domingo",
    2: "Segunda-feira",
    3: "Terça-feira",
    4: "Quarta-feira",
    5: "Quinta-feira",
    6: "Sexta-feira",
    7: "Sábado"
}

if numero in dias:
    print(f"O dia correspondente é {dias[numero]}.")
else:
    print("Número inválido! Digite um número entre 1 e 7.")
```

Peça um ano e informe se ele é bissexto. Crie um algoritmo em Python que verifique se o ano é divisível por 4 e não por 100, exceto se também for divisível por 400.

```python
ano = int(input("Digite um ano: "))
if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print("O ano é bissexto.")
else:
    print("O ano não é bissexto.")
```

Peça a quantidade de horas trabalhadas e o valor por hora. Se trabalhou até 40 horas, calcule o salário normalmente (horas × valor por hora). Se trabalhou mais de 40 horas, as horas extras devem ter um adicional de 50% (horas extras × valor por hora × 1.5). Crie um algoritmo em Python que realize esse cálculo corretamente.

```python
horas = float(input("Digite a quantidade de horas trabalhadas: "))
valor_hora = float(input("Digite o valor por hora: "))

if horas <= 40:
    salario = horas * valor_hora
else:
    horas_extras = horas - 40
    salario = (40 * valor_hora) + (horas_extras * valor_hora * 1.5)

print(f"O salário total é R$ {salario:.2f}")
```
